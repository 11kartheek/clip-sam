```python
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image

device = "cuda" if torch.cuda.is_available() else "cpu"
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
model.to(device)
```

```python
def load_and_preprocess_image(image_path):
    image = Image.open(image_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt")
    return inputs["pixel_values"].to(device)
```

```python
def encode_image(image_tensor):
    with torch.no_grad():
        image_features = model.get_image_features(image_tensor)
    return image_features

def encode_text(text):
    inputs = processor(text=[text], return_tensors="pt", padding=True)
    with torch.no_grad():
        text_features = model.get_text_features(**inputs.to(device))
    return text_features
```

```python
def get_similarity(image_features, text_features):
    image_features /= image_features.norm(dim=-1, keepdim=True)
    text_features /= text_features.norm(dim=-1, keepdim=True)
    similarity = ( image_features @ text_features.T).softmax(dim=-1)
    return similarity
```

```python
def main():
    image_path = "car.jpg"
    image_tensor = load_and_preprocess_image(image_path)
    image_features = encode_image(image_tensor)

    questions_answers = {
        "What color is the car?": "The car is red.",
        "What is the boy doing?": "The boy is playing with a ball.",
        "Is there a tree in the picture?": "Yes, there is a tree.",
        # Add more questions and answers as needed
    }

    print("Ask a question about the image (type 'exit' to quit):")

    while True:
        question = input("Question: ")
        if question.lower() == 'exit':
            break

        if True:#question in questions_answers:
            text_features = encode_text(question)
            similarity = get_similarity(image_features, text_features)
            print(similarity)
            if similarity.item() > 0.9:  # Adjust threshold as needed
                print("Answer:",similarity)
            else:
                print("I'm not sure about that.")
        else:
            print("I don't know the answer to that question.")

if __name__ == "__main__":
    main()
```
